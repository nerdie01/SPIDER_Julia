```julia
using JSON

dts::Vector{Float64} = round.(LinRange(0.001, 0.005, 9), sigdigits=5)

orig = JSON.parse(read("settings_copy.json"))

for dt ∈ dts
    orig["library_terms"]["∂^2u/∂t^2"]["scale"] = dt
    orig["library_terms"]["∂^2v/∂t^2"]["scale"] = dt
    orig["library_terms"]["∂^2Bx/∂t^2"]["scale"] = dt
    orig["library_terms"]["∂^2By/∂t^2"]["scale"] = dt

    write("settings-" * repr(dt) * ".json", JSON.json(orig))
end
```
