```julia
include("fdm/MHD_FDM.jl")
using .MHD_FDM, Random

include("spectral/MHDSpectral.jl")
using .MHDSpectral

using HDF5

N::Int64 = 64
steps::Int = 200
truncate = 101
dt = 0.0075

fω(x,y) = cos(x) + cos(y)
fj(x,y) = 2cos(2x) + cos(y)

forcing(x,y) = cos.(x) - cos.(y)

ν = 2^-6
η = 2^-6

ibutcher_A = hcat(1)
ibutcher_b = [1]

data_fdm = Dict()
data_spec = Dict()
data_fdm["x"], data_fdm["y"], data_fdm["u"], data_fdm["v"], data_fdm["Bx"], data_fdm["By"], data_fdm["p"] = MHD_FDM.mhd_fdm(fω, fj, N, dt, ibutcher_A, ibutcher_b, steps, ν, η, false, true, "fdm/plots/", forcing)
data_spec["x"], data_spec["y"], data_spec["u"], data_spec["v"], data_spec["Bx"], data_spec["By"], data_spec["p"] = MHDSpectral.mhd_spectral(fω, fj, N, dt, ibutcher_A, ibutcher_b, steps, ν, η, false, true, "spectral/plots/", forcing)

data_spec["x"] = reshape(data_spec["x"], 1, N)
data_spec["y"] = reshape(data_spec["y"], 1, N)

data_fdm["x"] = reshape(data_fdm["x"], 1, N)
data_fdm["y"] = reshape(data_fdm["y"], 1, N)

for key ∈ ("u", "v", "Bx", "By", "p")
    data_spec[key] = data_spec[key][:,:,truncate:end]
    data_fdm[key] = data_fdm[key][:,:,truncate:end]
end

data_fdm["t"] = collect(0:(steps - truncate - 1))*dt
data_spec["t"] = collect(0:(steps - truncate - 1))*dt

fid = h5open("data_fdm_jl.hdf5", "w")
for (key, value) ∈ data_fdm
    fid["result/value/" * key * "/value"] = value
end

close(fid)

fid = h5open("data_spec_jl.hdf5", "w")
for (key, value) ∈ data_spec
    fid["result/value/" * key * "/value"] = value
end

close(fid)
```
